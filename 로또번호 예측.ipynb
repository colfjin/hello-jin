```python
import pandas as pd
#로또 데이터 로드
lotto =pd.read_csv("./data/lotto_869.csv",encoding="euc-kr")
lotto=lotto.set_index("추첨일")
#날씨 데이터 로드
weather =pd.read_csv("./data/weater_seoul_200212_20190727.csv",encoding="euc-kr")
weather=weather.set_index("일시")
```

```python
#로또 데이터 확인
lotto.head()
```

```python
#날씨 데이터 확인
weather.head()
```

필요한 컬럼만 추출

```python
lotto=lotto.loc[:,["No_1","No_2","No_3","No_4","No_5","No_6","No_Bonus"]]
lotto.head()
```

```python
weather=weather.loc[:,["평균기온","평균 상대습도","평균 현지기압"]]
weather.head()
```

```python
#join="inner" : 로또 추첨일 날씨만 합하기 위해서
lotto_weather=pd.concat([lotto,weather],axis=1,join="inner")
lotto_weather.head()
```

```python
lotto_weather.shape
```

해당 번호가 1등 번호로 표시

```python
lotto_weather["등수"]=1
lotto_weather.head()
```

2등 번호와 등수를 추출해서 새로운 행에 추가(6개의 숫자 중 하나를 보너스번호로 변경)

```python
lotto_weather3=lotto_weather.copy()
for i in range(len(lotto_weather)):
    #해당 회차에서 각 번호 (6개) 대신에 하나씩 보너스 번호로 변경->2등
    
    for j in range(6):
        #한 줄씩 읽어서 lotto_weather2 저장
        lotto_weather2=lotto_weather.iloc[[i],:]
        #해당 줄에서 각 번호를 하나씩 보너스 번호로 교체
        lotto_weather2.iloc[:,j]=lotto_weather2.loc[:,"No_Bonus"]
        #등수 추가
        lotto_weather2.loc[:,"등수"]=2
        
        #추가된 데이터를 lotto_weather3의 행에 추가
        lotto_weather3=pd.concat([lotto_weather3,lotto_weather2])
        
        
```

```python
lotto_weather3.tail()
```

3등,4등,5등 데이터도 추가
- 3등 : 6개 자리수 중에서 1개 틀린것(5개가 맞은 것)
- 4등 : 6개 자리수 중에서 2개 틀린것(4개가 맞은 것)
- 5등 : 6개 자리수 중에서 3개 틀린것(3개가 맞은 것)

```python
lotto_weather2=lotto_weather.iloc[[4],:]
lotto_weather2.iloc[0,2]
```

```python
lotto_weather.iloc[0,2]
```

```python
lotto_weather3=lotto_weather.copy()
for i in range(len(lotto_weather)):
    
    for j in range(6):
        for k in range(1,46):
            lotto_weather2=lotto_weather.iloc[[i],:]
            lotto_weather2.iloc[0,j]=k
            num=lotto_weather.iloc[i,j]
            if k!=num:
                
        #등수 추가
                lotto_weather2.loc[:,"등수"]=3
        
        #추가된 데이터를 lotto_weather3의 행에 추가
                lotto_weather3=pd.concat([lotto_weather3,lotto_weather2])
```

```python
lotto123=pd.read_csv("./data/lotto_3degree.csv")
lotto123.head()
```

```python
lotto123.shape
```

```python
lotto123.tail()
```

```python
#보너스 번호 삭제
del lotto123["No_Bonus"]

lotto123.tail(10)
```

```python
X=lotto123.iloc[:,:-1]
y=lotto123.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=45)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
y_train_en=pd.get_dummies(y_train)
y_test_en=pd.get_dummies(y_test)
```

```python
y_train_en.shape,y_test_en.shape
```

로또 번호로 등수 예측 모델 만들기

```python
import numpy as np
import tensorflow as tf

seed=0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
lotto_model=Sequential()
lotto_model.add(Dense(45,input_dim=9,activation="relu"))
lotto_model.add(Dense(81,activation="relu"))
lotto_model.add(Dense(324,activation="relu"))
lotto_model.add(Dense(81,activation="relu"))
lotto_model.add(Dense(45,activation="relu"))
lotto_model.add(Dense(3,activation="softmax"))
lotto_model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
lotto_model.fit(X_train,y_train_en,epochs=10,batch_size=150)
```

```python
lotto_model.evaluate(X_test,y_test_en)
```
