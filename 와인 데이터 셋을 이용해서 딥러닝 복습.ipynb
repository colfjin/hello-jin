#### 와인 데이터 셋을 이용해서 딥러닝 복습

```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
wine=pd.read_csv("./data/wine.csv")
wine.head()
```

```python
wine.shape
```

```python
#데이터가 많았을 경우 일부 데이터만 샘플링해서 사용
wine=wine.sample(frac=0.6) #전체 데이터 중에서 60%만 사용

wine.shape
```

```python
#alcihol -> 회귀
#quality -> 다중
#type -> 이중분류
```

```python
# X, y
#회귀용 데이터 (y가 alcohol 컬럼)
y1=wine['alcohol']
X1=wine.drop(['alcohol'],axis=1)

#이진분류용 데이터 (y가 type 컬럼)
y2=wine['type']
X2=wine.drop(['type'],axis=1)

#다중분류용 데이터 (y가 quality 컬럼)
y3=wine['quality']
X3=wine.drop(['quality'],axis=1)
```

```python
seed=7
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
```

```python
y3.unique()
```

```python
len(y3)
```

```python
y3.reset_index(inplace=True,drop=True)
```

```python
X3.reset_index(inplace=True,drop=True)
```

```python
y=[]
for i in range(len(y3)):
    if y3[i]==6:
        y.append([1,0,0,0,0,0,0])
    elif y3[i]==5:
        y.append([0,1,0,0,0,0,0])
    elif y3[i]==7:
        y.append([0,0,1,0,0,0,0])
    elif y3[i]==4:
        y.append([0,0,0,1,0,0,0])
    elif y3[i]==3:
        y.append([0,0,0,0,1,0,0])
    elif y3[i]==8:
        y.append([0,0,0,0,0,1,0])
    elif y3[i]==9:
        y.append([0,0,0,0,0,0,1])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X3,y)
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()

#입력층
model.add(Dense(144,input_dim=12,activation="relu"))

#은닉층 추가

model.add(Dense(288,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(144,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(49,activation="relu"))
model.add(Dropout(0.2))
#출력층
model.add(Dense(7,activation="softmax"))
model.summary()
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["acc"])
```

```python
model.fit(X_train,y_train,epochs=1000,batch_size=10)
```

```python
model.evaluate(X_test,y_test)
```
