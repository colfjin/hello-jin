```python
#원 핫 인코딩(라벨인코딩 -> 원핫인코딩)
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

e=LabelEncoder()
e.fit(iris.target)

y=e.transform(iris.target)
```

```python
import pandas as pd
```

```python
iris_df=pd.DataFrame(data=iris.data,
                    columns=iris.feature_names)
iris_df
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris_df,y_en)
```

```python
model=Sequential()
model.add(Dense(16,input_dim=4,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dense(32,activation="relu"))
model.add(Dense(3,activation="softmax"))
model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["acc"])
```

```python
model.fit(X_train,y_train,epochs=100,batch_size=10)
```

```python
model.evaluate(X_test,y_test)
```

딥러닝은 교차검증을 어떻게?

```python
from sklearn.model_selection import KFold    #몇 개로 구분할 것인지 설정
from sklearn.model_selection import cross_val_score
```

```python
def deep_model():
    model=Sequential()
    model.add(Dense(16,input_dim=4,activation="relu"))
    model.add(Dense(64,activation="relu"))
    model.add(Dense(32,activation="relu"))
    model.add(Dense(3,activation="softmax"))
    model.compile(loss="categorical_crossentropy",
                  optimizer="adam",
                  metrics=["acc"])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
#KerasClassifier(딥러닝모델함수,epochs=반복수,batch_size=데이터수)
#분류학습
model=KerasClassifier(build_fn=deep_model,epochs=50,batch_size=10)
#몇개로 구분할 것인지
fold=KFold(n_splits=5,shuffle=True,random_state=0)

score=cross_val_score(model,X_test,y_test,cv=fold)
score
```

```python
score.mean()
```

```python
fold=KFold(n_splits=7,shuffle=True,random_state=0)

score=cross_val_score(model,X_test,y_test,cv=fold)
score
```

```python
score.mean()
```

베스트모델 찾아서 저장하기

```python
#피마 인디언 당뇨병 데이터
pima=pd.read_csv("./data/pima.csv",header=None,
                names=["pregnant", "plasma", "pressure", "thickness",
                       "insulin", "BMI", "pedigree", "age", "class"])
pima.head()
```

```python
X=pima.loc[:,:"age"]
y=pima['class']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
pima_model=Sequential()
pima_model.add(Dense(16,input_dim=8,activation="relu"))
pima_model.add(Dense(32,activation="relu"))
pima_model.add(Dense(64,activation="relu"))
pima_model.add(Dense(32,activation="relu"))
pima_model.add(Dense(16,activation="relu"))
pima_model.add(Dense(1,activation="sigmoid"))
pima_model.compile(loss="binary_crossentropy",
                  optimizer="adam",
                  metrics=["acc"])
```

```python
pima_model.fit(X_train,y_train,epochs=500,batch_size=10)
```

```python
pima_model.evaluate(X_test,y_test)
```

- 베스트 모델을 찾아서 저장하기 : ModelCheckpoint
- 더이상 학습이 나아지지 않으면 학습을 중단 : EarlyStopping

```python
import os
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
#모델을 저장할 폴더명
MODEL_FOLDER="./model/"
#해당 폴더가 없다면 해당 폴더를 생성
if not os.path.exists(MODEL_FOLDER):
    os.mkdir(MODEL_FOLDER)
    
#저장할 파일명 설정
#{epoch:04d} : 반복수를 4자리로 표시
#{acc:.4f} : 정확도를 소수점 4째자리까지 표시
modelpath = MODEL_FOLDER+"pima-{epoch:04d}-{val_acc:.4f}.hdf5"

#베스트를 찾아서 해당파일명으로 저장
#save_best_only : 더 나은 결과값만 저장
#ModelCheckpoint(filepath=파일패스,monitor=기준값,save_best_only=True/False)
mc=ModelCheckpoint(filepath=modelpath,
                  monitor="val_acc",
                   #회귀일 경우 monitor="loss"를 사용
                  save_best_only=True,
                  verbose=1)
# EarlyStopping(monitor=기준값, patience=조금 더 기다리는 횟수)
# patience=20 : 학습이 더 나아지지 않더라도 20회는 더 반복해줌
es = EarlyStopping(monitor="val_acc",patience=20)

#학습
history=pima_model.fit(X_train,y_train,epochs=50000,batch_size=50,
                      validation_split=0.33,
                      callbacks=[mc,es])
```

결과 시각화

```python
import numpy as np
```

```python
acc=history.history['acc']
val_acc=history.history['val_acc']

acc_len=np.arange(len(acc))

plt.plot(acc_len,acc,c="red")
plt.plot(acc_len,val_acc,c="blue")
```

```python
import pandas as pd
sonar=pd.read_csv("./data/sonar.csv",header=None)
sonar.head()
```

```python
#y의 클래스 수
sonar[60].unique()
```

```python
X=sonar.iloc[:,:60]
y=sonar.iloc[:,60]
X.shape,y.shape
```

```python
y_en=pd.get_dummies(y)
y_en.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y_en,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
import numpy as np
import tensorflow as tf
seed=0
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
sonar_model=Sequential()
sonar_model.add(Dense(120,input_dim=60,activation="relu"))
sonar_model.add(Dense(240,activation="relu"))
sonar_model.add(Dense(360,activation="relu"))
sonar_model.add(Dense(240,activation="relu"))
sonar_model.add(Dense(120,activation="relu"))
sonar_model.add(Dense(2,activation="softmax"))
sonar_model.compile(loss="categorical_crossentropy",
                  optimizer="adam",
                  metrics=["acc"])
#여기 데이터는 결과값이 2개이므로 loss 에 binary_crossentropy사용도 가능하다
```

```python
#validation_data=(X_test,y_test) : 테스트 세트를 선택
history2=sonar_model.fit(X_train,y_train,epochs=300,batch_size=10,
                        validation_data=(X_test,y_test))
```

```python
sonar_model.evaluate(X_test,y_test)
```

```python
import matplotlib.pyplot as plt
acc=history2.history['acc']
val_acc=history2.history['val_acc']

acc_len=np.arange(len(acc))

plt.plot(acc_len,acc,c="pink")
plt.plot(acc_len,val_acc,c="gold")
```

```python
sonar_model2=Sequential()
sonar_model2.add(Dense(120,input_dim=60,activation="relu"))
sonar_model2.add(Dropout(0.2))
sonar_model2.add(Dense(240,activation="relu"))
sonar_model2.add(Dropout(0.2))
sonar_model2.add(Dense(360,activation="relu"))
sonar_model2.add(Dropout(0.2))
sonar_model2.add(Dense(240,activation="relu"))
sonar_model2.add(Dropout(0.2))
sonar_model2.add(Dense(120,activation="relu"))
sonar_model2.add(Dropout(0.2))
sonar_model2.add(Dense(2,activation="softmax"))
sonar_model2.compile(loss="categorical_crossentropy",
                  optimizer="adam",
                  metrics=["acc"])
```

```python
history3=sonar_model2.fit(X_train,y_train,epochs=300,batch_size=10,
                        validation_data=(X_test,y_test))
```

```python
acc=history3.history['acc']
val_acc=history3.history['val_acc']

acc_len=np.arange(len(acc))

plt.plot(acc_len,acc,c="pink")
plt.plot(acc_len,val_acc,c="gold")
```

```python
print("훈련점수 : ", sonar_model.evaluate(X_train,y_train))
print("테스트점수 : ", sonar_model.evaluate(X_test,y_test))
```

```python
print("훈련점수 : ", sonar_model2.evaluate(X_train,y_train))
print("테스트점수 : ", sonar_model2.evaluate(X_test,y_test))
```
