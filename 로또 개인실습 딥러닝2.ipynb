```python
import pandas as pd
```

```python
lotto_data=pd.read_csv("./lotto_jin1",index_col=0)
lotto_data
```

```python
X=lotto_data.iloc[:,:-1]
y=lotto_data.iloc[:,-1]
```

```python
y_en=pd.get_dummies(y)
y_en
```

```python
X.shape,y_en.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
lotto_model=Sequential()
lotto_model.add(Dense(16,input_dim=8,activation="relu"))
lotto_model.add(Dense(64,activation="relu"))
lotto_model.add(Dense(128,activation="relu"))
lotto_model.add(Dense(128,activation="relu"))
lotto_model.add(Dense(64,activation="relu"))
lotto_model.add(Dense(45,activation="softmax"))
lotto_model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
lotto_model.fit(X,y_en,epochs=50,batch_size=10,
               validation_split=0.2)
```
