동영상 녹화하기

```python
import cv2

try:
    print("카메라 구동 시작")
    #cap=cv2.VideoCapture(0)
    cap=cv2.VideoCapture("./data/movie_test.mp4")
#예외 발생시 처리
except:
    print("카메라 구동 실패")

#녹화 설정
fps=30.0       #녹화할 프레임 수
width=int(cap.get(3))     #녹화할 영상의 가로 크기
height=int(cap.get(4))     #녹화할 영상의 세로 크기

#코덱설정 ->  MPEG
#fcc=cv2.VideoWriter_fourcc(*"MPEG")
fcc=cv2.VideoWriter_fourcc(*"DIVX")     #AVI
#fcc=cv2.VideoWriter_fourcc("D","I","V","X")   이렇게 작동도 가능

#녹화 설정 적용(파일명,코덱,프레임수,크기)
#out=cv2.VideoWriter("cctv01.mp4",fcc,fps,(width,height))
out=cv2.VideoWriter("cctv01.avi",fcc,fps,(width,height))
print("녹화를 시작합니다")

record=False
#카메라로부터 동영상 프레임을 계속 읽어옴
while True:
    #한 프레임씩 읽어서 frame에 저장하고 성공여부를 ret에 저장
    #성공 ret는 True,실패 ret는 False
    ret,frame=cap.read()
    
    #프레임 읽기 실패한 경우
    if not ret:
        print("프레임 읽기 실패")
        cap.release()     #카메라 해제
        cv2.destroyAllWindows()
        break
    
    #프레임 읽기 성공한 경우 동영상 저장
    cv2.imshow("cctv",frame)
    out.write(frame)
    
    #초당 30장의 프레임이 실행되도록 설정
    #사용자 키 입력 받기를 대기
    k=cv2.waitKey(33)
    
    if k==49:     #숫자 1이라면
        print("녹화를 종료합니다")
        cap.release()
        out.release()
        cv2.destroyAllWindows()
        break
        
    #숫자 2를 누르면 녹화가 시작되도록
    elif k==50:
        record=True
        
    if record==True:
        out.write(frame)
```

HSV 색상공간을 이용해서 원하는 색상만 추출하기
- RGB에서는 빛에 따라 색상 인식이 달라질수 도 있음


```python
import cv2
import numpy as np

img =cv2.imread("./data/colorballoon.jpg")
cv2.imshow("original",img)

#BGR색상을 HSV색상으로 변환
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#색상의 범위 (HSV)
#H(색상) :-180 ~ 180
#S(채도) : 0 ~ 255
#V(명도) : 0 ~ 255

#빨간색 영역
#최소값
lower=np.array([-10,100,100])
#최대값
upper=np.array([10,255,255])

#최소값과 최대값에 해당되는 범위를 마스크로 생성
mask=cv2.inRange(hsv,lower,upper)
#마스크에 포함된 이미지 픽셀값만을 (mask와 이미지를 and 연산)이미지에 표시
img2=cv2.bitwise_and(img,img,mask=mask)

cv2.imshow("red balloon",img2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 침식과 팽창

- 이진이미지 처리 방법
- Erosion(침식) : 이미지 경계부분에서 1을 0으로 변환하는 방법
- Dilation(팽창) : 이미지 경계부분에서 0을 1로 변환하는 방법
- iteration : 변환을 몇번 할 것인지 설정

```python
import cv2
import numpy as np

#흑백 이미지로 읽기 (cv2.)
img=cv2.imread("./data/number.png",cv2.IMREAD_GRAYSCALE)

cv2.imshow("original",img)

#침식과 팽창의 크기를 설정
kernel=np.ones((3,3),np.uint8)

#침식
erosion=cv2.erode(img,kernel,iterations=1)

cv2.imshow("erosion",erosion)

#팽창
dilation=cv2.dilate(img,kernel,iterations=1)

cv2.imshow("dilation",dilation)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

침식과 팽창을 이용해서 솔트잡음을 제거
- Openning : 침식 후에 팽창 연산을 수행(검정 배경에 흰색 잡음이 있는 경우)
- Closing : 팽창 후에 침식 연산을 수행(흰색 글씨에 검정색 잡음이 있는 경우)

```python
import cv2
import numpy as np
img = cv2.imread("./data/salt1.png",cv2.IMREAD_GRAYSCALE)

cv2.imshow("original",img)

#Openning
kernel= np.ones((3,3),np.uint8)

openning=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
cv2.imshow("openning",openning)

cv2.waitKey(0)
cv2.destroyAllWindows()

#안지워 진거 같으면 (3,3)->(5,5)등 수치로 올려주면 된다
```

### 탬플릿 매칭
- 큰 이미지 내에 작은 이미지 (탬플릿)가 있는 위치를 검색

```python
import cv2
import numpy as np

img1=cv2.imread("./data/game.png")
img2=cv2.imread("./data/apple.png",cv2.IMREAD_GRAYSCALE)

img1_gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)

#탬플릿 이미지의 크기 계산
#w : 가로 값, h : 세로 값
w,h = img2.shape[::-1]

#탬플릿 매칭
#cv2.TM_CCOEFF_NORMED : 매칭 방법
res=cv2.matchTemplate(img1_gray,img2,cv2.TM_CCOEFF_NORMED)

#매칭도가 기준값 이상인 위치만 가져온다(70%)
loc=np.where(res>=0.7)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img1,pt,(pt[0]+w,pt[1]+h),(0,0,255),2)
    


cv2.imshow("match",img1)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Hough변환을 이용해서 이미지에 원을 찾기

```python
import cv2
import numpy as np

img=cv2.imread("./data/findfigure.jpg")
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#허프 변환으로 원 찾기
#cv2.HoughCircles(이미지,검색방법,원의 중심점간의 최소거리,?,param1=Canny연산자의 최대값,param2=카운트값(작으면 많은 원을 찾음),minRadius=원의 최소반지름,maxRadius=원의 최대반지름)
#값을 변환해주는건 원의 중심점간의 최소거리, minRadius, maxRadius
circle = cv2.HoughCircles(img_gray,cv2.HOUGH_GRADIENT,1,10,
                         param1=60, param2=50,
                         minRadius=10, maxRadius=100)

if circle is not None:
    #반올림(중심좌표X,Y,반지름)
    circle=np.uint16(np.around(circle))
    
    #검색된 원을 그려준다
    for i in circle[0,:]:
        cv2.circle(img, (i[0],i[1]),i[2],(100,100,255),4)
        
    cv2.imshow("circle",img)
    
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("원이 없습니다")
```

### 모서리 (corner) 검출

```python
import cv2
import numpy as np

img=cv2.imread("./data/coner.png")
imgray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
#cv2.goodFeaturesToTrack(이미지,코너 갯수,기준값,코너점 간의 최소 거리)
corners=cv2.goodFeaturesToTrack(imgray,25,0.01,10)
#죄표값이므로 정수로 변환
corners=np.int0(corners)

#코너 좌표에 점을 찍어준다
for i in corners:
    #ravel() : 다차원 배열을 1차원 배열로 변환(리스트,튜플,배열 모두 사용가능)
    #flatten() : 다차원 배열을 1차원 배열로 변환(배열만 사용 가능)
    x,y=i.ravel()
    #-1 : 원의 내부의 색을 채움
    cv2.circle(img,(x,y),3,(100,100,255),-1)

cv2.imshow("corner",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

#얼굴 검출기
face_cascade=cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
#얼굴 안에 있는 눈 검출기
eye_cascade=cv2.CascadeClassifier("./data/haarcascade_eye.xml")

#얼굴과 눈을 검출할 이미지 로드
img=cv2.imread("./data/faces3.jpg")
imggray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#얼굴 검출기 적용해서 얼굴 좌표를 가져옴(시작x좌표, 시작y좌표, 길이,높이)
#face_cascade.dectectMultiScale(이미지명,스케일값,최소이웃값)
faces=face_cascade.detectMultiScale(imggray,1.3,5)

#찾은 얼굴 좌표를 이용해서 원이미지에 사각형을 그려줌
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(100,100,255),2)
    #얼굴 부분에서 눈 영역을 찾아서 표시
    #얼굴 ROI를 저장
    roi_gray=imggray[y:y+h,x:x+w]
    roi_color=img[y:y+h,x:x+w]
    
    # 눈 영역을 검출
    eyes=eye_cascade.detectMultiScale(roi_gray,1.1,2)
    
    #눈 영역을 이미지에 표시
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(100,100,50),2)
    
cv2.imshow("face",img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import pafy

url = "https://youtu.be/46FLsmUormA" # 유튜브 영상 url
vPafy = pafy.new(url)
play = vPafy.getbest(preftype="mp4")
#얼굴 검출기
face_cascade=cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
#얼굴 안에 있는 눈 검출기
eye_cascade=cv2.CascadeClassifier("./data/haarcascade_eye.xml")

# 카메라 연결
try:
    print("카메라 연결 성공")
    cap=cv2.VideoCapture(play.url)
except:
    print("카메라 연결 실패")

while True:
    ret,frame= cap.read()
    
    #동영상에서 한 장의 이미지를 읽는 것을 실패했다면
    if not ret:
        cap.release()
        cv2.destroyAllWindows()
        break
        
    #얼굴과 눈을 검출할 이미지 로드
    imggray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    #얼굴 검출기 적용해서 얼굴 좌표를 가져옴(시작x좌표, 시작y좌표, 길이,높이)
    #face_cascade.dectectMultiScale(이미지명,스케일값,최소이웃값)
    faces=face_cascade.detectMultiScale(imggray,1.3,5)

    #찾은 얼굴 좌표를 이용해서 원이미지에 사각형을 그려줌
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(100,100,255),2)
        #얼굴 부분에서 눈 영역을 찾아서 표시
        #얼굴 ROI를 저장
        roi_gray=imggray[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]

        # 눈 영역을 검출
        eyes=eye_cascade.detectMultiScale(roi_gray,1.3,2)

        #눈 영역을 이미지에 표시
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(100,100,50),2)

    cv2.imshow("face",frame)

    k=cv2.waitKey(33)
    if k==49:    #숫자 1을 누른 경우
        cap.release()
        cv2.destroyAllWindows()
        break
```
