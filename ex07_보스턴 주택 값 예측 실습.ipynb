```python
from sklearn.linear_model import LinearRegression
```

```python
#y 데이터,boston의 target값 사용
house_price=boston.target
```

### train과 test로 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(boston_df,house_price)
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
#copy
boston_copy=X_train.copy()#copy로 해놓으면 X_train값이 바뀌는 일이 없다
boston_copy['Price']=y_train
boston_copy
```

```python
boston_copy.corr()
```

```python
#특성 확장
#칼럼값을 곱해서 새로운 칼럼을 만들자
extend_X_train=X_train.copy()
for col1 in X_train.columns:
    for col2 in X_train.columns:
        extend_X_train[col1+"x"+col2]=X_train[col1]*X_train[col2]
```

```python
extend_X_train.shape
```

```python
extend_X_test=X_test.copy()
for col1 in X_test.columns:
    for col2 in X_test.columns:
        extend_X_test[col1+"x"+col2]=X_test[col1]*X_test[col2]

extend_X_test.shape

```

### 1. 셀 합치기
- edit모드에서 shift +m

### 2. 셀 나누기
- command모드에서 ctrl+shift+-

### Lasso

```python
lasso=Lasso(alpha=1)
lasso.fit(extend_X_train,y_train)
print(lasso.score(extend_X_train,y_train))
print(lasso.score(extend_X_test,y_test))
print(np.sum(lasso.coef_!=0))#사용한 특성의 개수
```

```python
#모든 특성의 가중치
lasso.coef_
```

```python
lasso.coef_!=0
```

```python
lasso=Lasso(alpha=100)
lasso.fit(extend_X_train,y_train)
print(lasso.score(extend_X_train,y_train))
print(lasso.score(extend_X_test,y_test))
print(np.sum(lasso.coef_!=0))#사용한 특성의 개수
#특성 사용갯수가 적어져버려서 과소적합
```

```python
lasso=Lasso(alpha=0.01)
lasso.fit(extend_X_train,y_train)
print(lasso.score(extend_X_train,y_train))
print(lasso.score(extend_X_test,y_test))
print(np.sum(lasso.coef_!=0))#사용한 특성의 개수
```

```python
lasso=Lasso(alpha=0.001)
lasso.fit(extend_X_train,y_train)
print(lasso.score(extend_X_train,y_train))
print(lasso.score(extend_X_test,y_test))
print(np.sum(lasso.coef_!=0))#사용한 특성의 개수
#특성 사용갯수가 너무 많아서 과대적합
```

```python
ridge=Ridge(alpha=1)
ridge.fit(extend_X_train,y_train)
print(ridge.score(extend_X_train,y_train))
print(ridge.score(extend_X_test,y_test))
print(np.sum(ridge.coef_!=0))
```
