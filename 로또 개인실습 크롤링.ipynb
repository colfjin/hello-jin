```python
turn_list=[]
year_list=[]
month_list=[]
day_list=[]
num1_list=[]
num2_list=[]
num3_list=[]
num4_list=[]
num5_list=[]
num6_list=[]
bonus_list=[]

for i in range(1,96,1):
    res=req.get("https://www.dhlottery.co.kr/gameResult.do?method=statByColor&sortOrder=DESC&startPage=1&endPage=10&currentPage="+str(i))
    lotto_data=bs(res.text,'lxml')
    turn=lotto_data.select('tr> td:nth-child(1)')
    date=lotto_data.select('tr> td:nth-child(2)')
    num1=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(1)')
    num2=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(2)')
    num3=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(3)')
    num4=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(4)')
    num5=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(5)')
    num6=lotto_data.select(' tr> td:nth-child(3)>span:nth-child(6)')
    bonus=lotto_data.select(' tr> td:nth-child(4)')
    
    for j in range(len(turn)):
        turn_list.append(int(turn[j].text))
        year_list.append(int(date[j].text.split('-')[0]))
        month_list.append(int(date[j].text.split('-')[1]))
        day_list.append(int(date[j].text.split('-')[2]))
        num1_list.append(int(num1[j].text))
        num2_list.append(int(num2[j].text))
        num3_list.append(int(num3[j].text))
        num4_list.append(int(num4[j].text))
        num5_list.append(int(num5[j].text))
        num6_list.append(int(num6[j].text))
        bonus_list.append(int(bonus[j].text))
```

```python
odd_list=[]
even_list=[]
total_list=[]

for i in range(1,96,1):
    res=req.get("https://www.dhlottery.co.kr/gameResult.do?method=statOddEven&sortOrder=DESC&startPage=1&endPage=10&currentPage="+str(i))
    lotto_data=bs(res.text,'lxml')
    odd=lotto_data.select('tbody > tr> td:nth-child(3)')
    even=lotto_data.select('tbody > tr> td:nth-child(4)')
    total=lotto_data.select('tbody > tr> td:nth-child(5)')
    
    for j in range(len(odd)):
        odd_list.append(len(odd[j].text.strip().split("\n")))
        even_list.append(len(even[j].text.strip().split("\n")))
        total_list.append(int(total[j].text))
```
