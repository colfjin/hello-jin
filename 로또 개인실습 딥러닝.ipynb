```python
X=lotto_data.iloc[:,:-1]
y=lotto_data.iloc[:,-1]
```

```python
y_en=pd.get_dummies(y)
y_en
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y_en,random_state=45)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
lotto_model=Sequential()
lotto_model.add(Dense(18,input_dim=9,activation="relu"))
lotto_model.add(Dense(72,activation="relu"))
lotto_model.add(Dense(144,activation="relu"))
lotto_model.add(Dense(36,activation="relu"))
lotto_model.add(Dense(18,activation="relu"))
lotto_model.add(Dense(2,activation="softmax"))
lotto_model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
lotto_model.fit(X_train,y_train,epochs=50,batch_size=50,
               validation_data=(X_test,y_test))
```
